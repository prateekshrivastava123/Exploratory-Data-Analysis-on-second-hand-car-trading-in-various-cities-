```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
brand = []
cleaned_data = []
car_names = []
model=[]
cost = []
prices = []
EMI= []
car_info = []
kilometers = []
variant = []
city = []

l=[1, 2, 4,5,7,8,9,10,16,70,43,23,33]
for i in l:
    url = 'https://www.cartrade.com/buy-used-cars/#so=-1&sc=-1&city=105'+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.text, "html.parser")
    a =soup.find_all('h2', class_="h2heading truncate") 
    for i in a:
        brand.append(i.text.strip())
    for i in brand:
        clean_entry = re.sub(r'\s*\[\d+-\d+\]', '', i)
        cleaned_data.append(clean_entry)
    for i in cleaned_data:
        match = re.findall(r'\d+\s(.+)', i)
        car_names.append(match)
    for i in a:
        text=i.text
        text= text.replace(',', '')
        regex=re.findall("\t(\d+)\s\w+",text)
        model.append(regex)
    b =soup.find_all('div', class_='cr_prc')
    for i in b:
        cost.append(i.text)
    for i in cost:
        price_match = re.findall('^₹(\d+.*)', i)
        prices.append(price_match)
    c =soup.find_all('div', class_="info_cr_new")
    for i in c:
        car_info.append(i.text)
    
        kilometers = re.findall(r'(\d[\d,]*) KMs', ''.join(car_info))
    
        variant = re.findall(r'\|(\w+)\|\w+\n', ''.join(car_info))
    
        city = re.findall(r'\|\w+\|(\w+)\n', ''.join(car_info))
    
    d =soup.find_all('div', class_="splLinks-wrap")
    for i in d:
        text=i.text
        text= text.replace(',', '')
        regex=re.findall("\s\\W+(\d+)\s",text)
        EMI.append(regex)
    
   
       
    
  
  
```
